bi-variate and multi-variate analysis

```python
plt.figure(figsize=(7,6))
sns.scatterplot(x='quantity',y='price',data=df,edgecolor='k')
plt.title('scatter plot of quantity vs price with density estimation')
plt.xlabel('quantity')
plt.ylabel('price')
plt.show()
```

```python
plt.figure(figsize=((8,6)))
sns.lineplot(data=df,x='quantity',y='price',color='blue')
plt.title('line plot eamxple')
plt.xlabel('x variable')
plt.ylabel('y variable')
plt.show()
```

```python
plt.figure(figsize=(7,6))
sns.histplot(df['quantity'],kde=True,color='blue',bins=30)
sns.histplot(df['price'],kde=True,color='red',bins=30)
plt.title('marginal histograms with density estimation')
plt.xlabel('value')
plt.ylabel('density')
plt.legend(['quantity','price'])
plt.show()
```

```python
sns.jointplot(x='quantity',y='price',data=df,kind='hex',cmap='viridis',height=6)
plt.show()
```
